```python
import os
import glob
import pandas as pd
import xml.etree.ElementTree as ET
import tensorflow as tf
from numpy import expand_dims
from tensorflow.keras.preprocessing import image
#from glob import glob
```

```python
from glob import glob
image_files = glob("data/images_new/val/testjpgs/*.jpg")
xml_files = glob("data/images_new/val/testxmls/*.xml")
images = [name for name in image_files]
xmls = [name for name in xml_files]
#for i in range(len(images)):
#  if (images[i][27:-4] != xml_files[i][27:-4]):
#    print(images[i][27:-4], xml_files[i][27:-4])
#image_files[0][27:-4]    
print(len(xmls), len(images))
```

```python
def xml_to_csv():
    xml_list = []
    #print('0')
    #print(path + '/*.xml')
    #print(glob.glob('/object_detection/test_images/*.xml'))
    for xml_file in glob.glob('data/images_new/val/testxmls/*.xml'):
        #print(xml_file)
        tree = ET.parse(xml_file)
        root = tree.getroot()
        flag = False
        #print('1')
        for member in root.findall('object'):
            value = (root.find('filename').text,
                     int(root.find('size')[0].text),
                     int(root.find('size')[1].text),
                     member[0].text,
                     int(member[4][0].text),
                     int(member[4][1].text),
                     int(member[4][2].text),
                     int(member[4][3].text)
                     )
            #print('2')
            #print(value)
            xml_list.append(value)
    column_name = ['filename', 'width', 'height', 'class', 'xmin', 'ymin', 'xmax', 'ymax']
    xml_df = pd.DataFrame(xml_list, columns=column_name)
    return xml_df


def main():
	
    image_path = os.path.join(os.getcwd(), '/data/images_new/val/testjpgs')
    xml_df = xml_to_csv()
    xml_df.to_csv('data/images_new/val/valtest_labels.csv', index=None)

    #image_path = os.path.join(os.getcwd(), '/data/images_new/train')
    #xml_df = xml_to_csv()
    #xml_df.to_csv('object_detection/data/test_labels.csv',index=None)

main()
```
